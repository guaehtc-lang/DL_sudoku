## 3. Importar librerías

````python
import shutil

from ultralytics import YOLO
````

## 4. Rutas

````python
import os

ruta_dataset = (
    "/content/drive/MyDrive/"
    "Colab Notebooks/img"
)

ruta_yaml = "/content/data_colab.yaml"

ruta_modelo_final = (
    "/content/drive/MyDrive/"
    "Colab Notebooks/best_yolo_sudoku.pt"
)

contenido_yaml = f"""
path: {ruta_dataset}

train: images/train
val: images/val
test: images/test

names:
  0: sudoku
"""

with open(
    ruta_yaml,
    "w",
    encoding="utf-8"
) as archivo:

    archivo.write(contenido_yaml)

print("Dataset:", ruta_dataset)
print("Modelo final:", ruta_modelo_final)
````

## 5. Entrenar el modelo

````python
modelo = YOLO("yolov8s.pt")

modelo.train(
    data=ruta_yaml,
    epochs=30,
    imgsz=640,
    batch=16,
    device=0,
    project="/content/runs_yolo",
    name="sudoku",
    exist_ok=True
)
````

## 6. Guardar el mejor modelo en Google Drive

````python
ruta_best = (
    "/content/runs_yolo/"
    "sudoku/weights/best.pt"
)

shutil.copy(
    ruta_best,
    ruta_modelo_final
)

print(
    "Modelo guardado en:",
    ruta_modelo_final
)
````
