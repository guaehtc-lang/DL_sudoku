```python
# Importamos las librerías

import pandas as pd
import numpy as np
```

```python
# Ruta del dataset

ruta_csv = "data/sudoku.csv"


# Cargamos una muestra de 9.000 Sudokus

df = pd.read_csv(
    ruta_csv,
    nrows=9000,
    dtype=str
)


print("Dataset cargado correctamente")
print("Dimensión:", df.shape)
```

```python
# Mostramos las primeras filas

df.head()
```

```python
df.columns.tolist()
```

```python
# Convertimos los Sudokus en matrices

X = np.array(
    [
        list(sudoku)
        for sudoku in df["puzzle"]
    ],
    dtype=np.int8
).reshape(-1, 9, 9)


y = np.array(
    [
        list(sudoku)
        for sudoku in df["solution"]
    ],
    dtype=np.int8
).reshape(-1, 9, 9)


print("Dimensión de X:", X.shape)
print("Dimensión de y:", y.shape)
```

```python
# Mostramos el primer Sudoku

print("PUZZLE")
print(X[0])

print("\nSOLUTION")
print(y[0])
```
